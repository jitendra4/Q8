```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

#Importing requests
import requests


# importing regex
import re
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException, ElementNotVisibleException, ElementNotInteractableException, SessionNotCreatedException, TimeoutException    
```

```python
#Lets first connect to the web drivers
driver=webdriver.Chrome(r"C:\Users\G\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
driver.get("https://www.imdb.com/list/ls095964455/")
time.sleep(3)
```

```python
#intialising lists
name=[]
year=[]
genre=[]
run_time=[]
rating=[]
vote=[]
```

```python
#Movie Name
names=driver.find_elements_by_xpath("//h3[@class='lister-item-header']")
for i in names:
    name.append(i.text)
```

```python
#year
years=driver.find_elements_by_xpath("//span[@class='lister-item-year text-muted unbold']")
for i in years:
    year.append(i.text)
```

```python
#Genre
genres=driver.find_elements_by_xpath("//span[@class='genre']")
for i in genres:
    genre.append(i.text)
```

```python
#Runtime
run_times=driver.find_elements_by_xpath("//span[@class='runtime']")
for i in run_times:
    run_time.append(i.text)
```

```python
#rating
ratings=driver.find_elements_by_xpath("//span[@class='ipl-rating-star__rating']")
for i in ratings:
    rating.append(i.text)
```

```python
#Vote
votes=driver.find_elements_by_xpath("//p[@class='text-muted text-small']")
for i in votes:
    vote.append(i.text)
```

```python
df=pd.DataFrame({})
df["names"]=name[0:20]
df["years"]=year[0:20]
df["genres"]=genre[0:20]
df["run_time"]=run_time[0:20]
df["ratings"]=rating[0:20]
df["vote"]=votes[0:20]
df

```
